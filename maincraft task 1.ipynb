```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\supar\Downloads\student\student-mat.csv", delimiter=';')

# --- Data Cleaning and Inspection ---
print("--- Missing Values Check ---")
print(df.isnull().sum())
print("\n")

print("--- Removing Duplicates ---")
initial_shape = df.shape
df.drop_duplicates(inplace=True)
final_shape_duplicates = df.shape
print(f"Initial shape: {initial_shape}")
print(f"Shape after removing duplicates: {final_shape_duplicates}")
print("\n")

print("--- Dataset Shape and Data Types ---")
print(f"Dataset shape: {df.shape}")
print("\n")
df.info()
print("\n")

# Convert grade columns to a numeric type for plotting and calculations
df['G1'] = pd.to_numeric(df['G1'])
df['G2'] = pd.to_numeric(df['G2'])
df['G3'] = pd.to_numeric(df['G3'])

# --- Data Visualization ---
# 1. Histogram of Grades (G1, G2, G3)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
df['G1'].hist(bins=10, edgecolor='black')
plt.title('Distribution of G1 Grades')
plt.xlabel('G1 Grade')
plt.ylabel('Frequency')
plt.grid(False)

plt.subplot(1, 3, 2)
df['G2'].hist(bins=10, edgecolor='black')
plt.title('Distribution of G2 Grades')
plt.xlabel('G2 Grade')
plt.ylabel('Frequency')
plt.grid(False)

plt.subplot(1, 3, 3)
df['G3'].hist(bins=10, edgecolor='black')
plt.title('Distribution of G3 Grades')
plt.xlabel('G3 Grade')
plt.ylabel('Frequency')
plt.grid(False)

plt.tight_layout()
plt.savefig('histogram_grades.png')
plt.show()

# 2. Scatter plot: Study Time vs. Final Grade (G3)
plt.figure(figsize=(8, 6))
plt.scatter(df['studytime'], df['G3'], alpha=0.6)
plt.title('Study Time vs. Final Grade (G3)')
plt.xlabel('Study Time (1: <2h, 2: 2-5h, 3: 5-10h, 4: >10h)')
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.savefig('studytime_vs_grades.png')
plt.show()

# 3. Bar chart: Male vs. Female Average Score (G3)
avg_scores = df.groupby('sex')['G3'].mean().reset_index()
plt.figure(figsize=(7, 5))
plt.bar(avg_scores['sex'], avg_scores['G3'], color=['skyblue', 'salmon'])
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.grid(axis='y', linestyle='--')
plt.savefig('male_vs_female_scores.png')
plt.show()
```

```python
Here’s the analysis from  dataset:

Average final grade (G3): ≈ 10.42

Students scoring above 15: 40 students

Correlation between study time and performance (G3): 0.098 (a very weak positive correlation)

Average grade by gender:

Female: 9.97

Male: 10.91

 Conclusion:

The average performance is around 10.4.

Only a small number of students (40) scored above 15.

Study time is only weakly correlated with performance.

On average, males slightly outperform females in the final grade.

```
